```python
def  getLimitsAndMedian(attribute):
  # Cuartil 1
  q1 = attribute.quantile(0.25)
 
  # Cuartil 3
  q3 = attribute.quantile(0.75)

  med = attribute.median()
  
  # IQR
  iqr = q3-q1
  
  upper_bound = q3+(1.5*iqr)
  lower_bound = q1-(1.5*iqr)

  print(q1, med, q3, iqr)
  return upper_bound, lower_bound, med
```

##### Tratamiento de outliers de Duracion del contrato (Dias)

```python
upper_bound, lower_bound,median = getLimitsAndMedian(clean_contratos_df['Duracion del contrato (Dias)'])
```

```python
clean_contratos_df.loc[(clean_contratos_df["Duracion del contrato (Dias)"] > upper_bound) | (clean_contratos_df["Duracion del contrato (Dias)"] < lower_bound),"Duracion del contrato (Dias)"]=median
```
